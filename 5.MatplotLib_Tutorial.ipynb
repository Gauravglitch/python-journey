# MatplotLib Tutorial

```python
import matplotlib.pyplot as plt
%matplotlib inline

import numpy as np
```

```python
## Simple examples

x=np.arange(0,10)
y=np.arange(11,21)
```

```python
a=np.arange(40,50)
b=np.arange(50,60)
```

```python
# plotting using matplotlib

#plt scatter

plt.scatter(x,y,c='g') #c='g' means color = green
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.title('Graph in 2D')
plt.savefig('Test.png'); #saves the figure

```

```python
#plt plot
y=x*x
plt.plot(x,y,'r*', linestyle='dashed', linewidth=2, markersize=12)
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.title('2D diagram');
```

```python
# creating subplots
plt.subplot(2,2,1)
plt.plot(x,y,'r--')
plt.subplot(2,2,2)
plt.plot(x,y,'g')
plt.subplot(2,2,3)
plt.plot(y,x,'b*-')
plt.subplot(2,2,4)
plt.plot(y,x,'bo');
```

```python
np.pi # value of pi 
x=np.arange(0,4*np.pi,0.1)
y=np.sin(x)
plt.title("Sine wave form")
plt.plot(x,y);
```

```python
#Bar plot
x1=[2,8,10]
y1=[11,16,9]

x2=[3,9,11]
y2=[6,15,4]

plt.bar(x1,y1)
plt.bar(x2,y2,color='g')
plt.xlabel("X axis")
plt.ylabel("Y axis")
plt.title("Bar Graph");
```

## Box plot using matplotlib

```python
data=[np.random.normal(0,std,100) for std in range(1,4)]

#rectangular box plot
plt.boxplot(data,vert=False,patch_artist=True);
```

```python
#Data to plot
labels ='Python','C++','Ruby','Java'
sizes=[215,130,245,210]
colors=['gold','yellowgreen','lightcoral','lightskyblue']
explode=(0.1,0,0,0) #explode 1st slice

#plot
plt.pie(sizes,explode=explode, labels=labels, colors=colors,
autopct='%1.1f%%', shadow=True)
plt.axis('equal');
```
