# Seaborn Tutorial

```python
import seaborn as sns
```

```python
df=sns.load_dataset("tips")
df.head()
```

### Correlation with Heatmap

```python
df.dtypes
```

```python
df.info()
```

```python
# Select specific columns
selected_columns = ['total_bill', 'tip', 'size']
df_selected = df[selected_columns]

# Calculate the correlation matrix
df_selected.corr()
```

```python
sns.heatmap(df_selected.corr())
```

## JoinPlot

Univariate Analysis

```python
sns.jointplot(x='tip',y='total_bill',data=df,kind='reg');
```

### PairPlot
also known as scatterplot

```python
sns.pairplot(df_selected)
```

```python
sns.pairplot(df,hue='sex')
```

```python
df["smoker"].value_counts()
```

### Distplot

```python
sns.distplot(df['tip'])
```

```python
sns.distplot(df['tip'], kde=False, bins=10)
```

## Catagorical Plot

- boxplot
- violinplot
- countplot
- barplot

```python
## Count plot
sns.countplot(x='day', data=df);
```

```python
## Barplot
sns.barplot(x='total_bill',y='sex',data=df)
```

```python
## Box plot
sns.boxplot(y='day',x='total_bill',data=df, palette='rainbow',hue='smoker')
```

```python
## Violin Plot
sns.violinplot(x='total_bill',y='day',data=df,palette='rainbow',hue='smoker');
```
